```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
'Student_ID': [1,2,3,4,5],
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
'Age': [25, None, 22,28,24],
'Grade' : ['A','B','C',None,'A'],
'City' : ['New York', 'San Fransico', 'Los Angeles','Chicago',
'Miami']
}

df = pd.DataFrame(data)
print("Task 1: Loaded Dataset")
print(df)

missing_cols = df.columns[df.isnull().any()]
missing_percentage = df[missing_cols].isnull().mean() * 100 
print("\nTask 2: Missing Values")
print("Columns with missing values:", list(missing_cols))
print("Percentage of missing values:\n", missing_percentage)

df['Age'].fillna(df['Age'].mean(), inplace=True) 
df['Grade'].fillna('Unknown', inplace=True) 
print("\nTask 3: Handled Missing Values")
print(df)

df['Name'] = df['Name'].str.upper()
print("\nTask 4: Converted Names to Uppercase")
print(df)

df['City_First_Letter'] = df['City'].str[0]
print("\nTask 5: Extracted First Letter of City")
print(df)

processed_file_path = "processed_dataset.csv"
df.to_csv(processed_file_path, index=False)
print("\nlask 6: Saved the Processed Dataset to",processed_file_path)
df['city_No_Spaces'] = df['City'].str.replace(' ','') 
print("\nTask 7: Replaced Spaces in city names")
print(df)

df ['Has_Substring_LA'] = df['City'].str.contains('LA') 
print("\nTask 8: Checked for Substring 'LA' in City Names")
print(df)

df[['First_name', 'Last_name']] = df['Name'].str.extract(r'(?P<first_name>\S+)(?:\s+(?P<Last_name>\S+))?',expand = True)
print("\nTask 9 : Extracted first and last name")
print(df)

df['Full_name'] = df['Name']
df = df.drop('Name', axis = 1) 
print("\nTask 10: Joined Firs5t and Last Name")
print(df)

df["Grade_Digits"] = df["Grade"].str.extract("(\d+)", expand=False)
print("\nTask 11: Extracted Digits from Grade")
print(df)

df["Vowel Count"] = df['City'].str.count("[aeiouAEIOU]")
print("\nTask 12: Counted Vowels in City hames")
print(df)

updated_file_path = 'updated_dataset.csv'
df.to_csv(updated_file_path, index=False)
print("\nTask 13: Saved the Updated Dataset to", updated_file_path)
```
